### üñ•Ô∏è Step-by-Step: Mac Setup.

1. Install Google Drive for Desktop

Download from: https://www.google.com/drive/download/
- Sign in with the same Google account you use for Colab.
- Choose to stream or mirror files (either works).

2. Access a Shared Folder '12 - DataShip Client Success Reporting' on Google Drive

If the project folder '12 - DataShip Client Success Reporting' was shared with you
- Go to https://drive.google.com/drive/shared-with-me
- Right-click the folder ‚Üí Add shortcut to Drive
- Choose a location inside My Drive

üí° This is crucial ‚Äî only folders inside "My Drive" will sync to your computer.

3. Find the Synced Folder Path

    Once the shortcut is created, Drive will sync it to:
    ```~/Library/CloudStorage/GoogleDrive-your_email@gmail.com/My Drive/```

    Check if the '12 - DataShip Client Success Reporting' folder exists via terminal:
    ```ls ~/Library/CloudStorage```

4. IMPORTANT: Avoid Double-Folder Issue When Cloning

    If the folder 12 - DataShip Client Success Reporting already exists in your Drive, do not clone the GitHub repo into it as a subfolder.

    Instead, do one of the following:

    ‚úÖ Option A: Treat the existing folder as your repo folder

        ```
        cd ~/Library/CloudStorage/GoogleDrive-your_email@gmail.com/My\ Drive/12\ -\ DataShip\ Client\ Success\ Reporting

        git init
        git remote add origin https://github.com/abualfaki/ClientSuccess-Dashboard.git
        # To get the latest code from GitHub
        git pull origin main # Optional if you want to sync with the latest Github Version
        ```

        ‚úÖ Best Practice:
        If possible, always git clone directly from GitHub instead of downloading from Google Drive.

    ‚úÖ Option B: Clone directly into the folder (only if it's empty or you're setting it up fresh)
        
        ```
        cd ~/Library/CloudStorage/GoogleDrive-your_email@gmail.com/My\ Drive/
        git clone https://github.com/abualfaki/ClientSuccess-Dashboard.git "12 - DataShip Client Success Reporting"
        ```

        ‚ùå Avoid running git clone inside the folder ‚Äî that would create a nested subfolder like 12 - DataShip Client Success Reporting/ClientSuccess-Dashboard, which can lead to confusion.

5. Commit & Push From Your Local Terminal

    After making changes (e.g. in Colab):

        ```
        git status
        git add .
        git commit -m "Update from Colab"
        git push origin main
        ```

### üö´ Protecting the main Branch (Best Practice for Teams)

When working in teams to connect GitHub to Google Drive, it‚Äôs important to protect the `` branch so it isn‚Äôt accidentally overwritten by someone else‚Äôs push.

#### ‚úÖ Option 1: Use Feature Branches

    Ask collaborators to create a branch for their work:

    ```
    git checkout -b feature-yourname
    # make changes
    git add .
    git commit -m "Your updates"
    git push origin feature-yourname
    ```

    Then, go to GitHub and create a Pull Request to merge into main.

#### ‚úÖ Option 2: Enable Branch Protection in GitHub

    To prevent direct pushes to main:

    Go to your GitHub repo ‚Üí Settings ‚Üí Branches

    Click Add rule under Branch protection rules

    Type main as the branch name

    Check:

    ‚úÖ Require pull request before merging

    ‚úÖ Require approvals (optional)

    This adds a safety net and promotes clean collaboration.
